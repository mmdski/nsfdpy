```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from nsfd.field import ScalarField


m_rows = 10
n_cols = 10

x_min = -5
x_max = 5
y_min = 0
y_max = 10

dx = (x_max - x_min) / (n_cols)
dy = (y_max - y_min) / (m_rows)

field_x_min = x_min + dx / 2
field_x_max = x_max - dx / 2
field_y_min = y_min + dy / 2
field_y_max = y_max - dy / 2

init_s_value = 1
s_field = ScalarField(m_rows, n_cols, field_x_min, field_x_max, field_y_min, field_y_max, init_s_value)
```

```python
fig, ax = plt.subplots(1, 1, dpi=200, figsize=(3,3))
ax = s_field.plot(ax)

v_grid_lines = np.linspace(x_min, x_max, n_cols + 1)
h_grid_lines = np.linspace(y_min, y_max, m_rows + 1)

ax.vlines(v_grid_lines, y_min, y_max, colors="gray", linestyles="dashed", linewidth=0.5)
ax.hlines(h_grid_lines, x_min, x_max, colors="gray", linestyles="dashed", linewidth=0.5)
ax.set_box_aspect(1)
```
