```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from nsfd.grid import PGrid, UGrid, VGrid


grid_props = {
    "m_rows": 10,
    "n_cols": 10,
    "x_min": -5,
    "x_max": 5,
    "y_min": 0,
    "y_max": 10,
}

p_grid = PGrid(grid_props)
u_grid = UGrid(grid_props)
v_grid = VGrid(grid_props)
```

```python
fig, ax = plt.subplots(1, 1, dpi=200, figsize=(3, 3))
ax = p_grid.plot(ax, c="r", s=0.5)
ax = u_grid.plot(ax, c="g", s=0.5)
ax = v_grid.plot(ax, c="b", s=0.5)

v_grid_lines = np.linspace(
    grid_props["x_min"], grid_props["x_max"], grid_props["n_cols"] + 1
)
h_grid_lines = np.linspace(
    grid_props["y_min"], grid_props["y_max"], grid_props["m_rows"] + 1
)

ax.vlines(
    v_grid_lines,
    grid_props["y_min"],
    grid_props["y_max"],
    colors="gray",
    linestyles="dashed",
    linewidth=0.5,
)
ax.hlines(
    h_grid_lines,
    grid_props["x_min"],
    grid_props["x_max"],
    colors="gray",
    linestyles="dashed",
    linewidth=0.5,
)
ax.set_box_aspect(1)
```
