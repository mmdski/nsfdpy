```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.path as mpath
import numpy as np

from nsfd.grid import PGridGeom, UGridGeom, VGridGeom

grid_props = {
    "imax": 10,
    "jmax": 10,
    "xlength": 100,
    "ylength": 100,
}

p_grid_geom = PGridGeom(grid_props)
u_grid_geom = UGridGeom(grid_props)
v_grid_geom = VGridGeom(grid_props)

ax = plt.axes()

ax = p_grid_geom.plot(ax, grid=True, boundary=True, c="b", s=0.5)
ax = u_grid_geom.plot(ax, c="r", s=0.5)
ax = v_grid_geom.plot(ax, c="g", s=0.5)
ax.set_aspect('equal', 'box')
```
