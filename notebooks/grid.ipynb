```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from nsfd.grid import Grid


m_rows = 10
n_cols = 10

x_min = -5
x_max = 5
y_min = 0
y_max = 10

dx = (x_max - x_min) / (n_cols)
dy = (y_max - y_min) / (m_rows)

p_x_min = x_min + dx / 2
p_x_max = x_max - dx / 2
p_y_min = y_min + dy / 2
p_y_max = y_max - dy / 2

init_p_value = 1
p_grid = Grid(m_rows, n_cols, p_x_min, p_x_max, p_y_min, p_y_max, init_p_value)

u_x_min = x_min + dx
u_x_max = x_max
u_y_min = y_min + dx / 2
u_y_max = y_max - dx / 2
u_grid = Grid(m_rows, n_cols, u_x_min, u_x_max, u_y_min, u_y_max)

v_x_min = x_min + dx / 2
v_x_max = x_max - dx / 2
v_y_min = y_min + dx
v_y_max = y_max
v_grid = Grid(m_rows, n_cols, v_x_min, v_x_max, v_y_min, v_y_max)
```

```python
fig, ax = plt.subplots(1, 1, dpi=200, figsize=(3,3))
ax = p_grid.plot(ax, s=0.5)
ax = u_grid.plot(ax, s=0.5)
ax = v_grid.plot(ax, s=0.5)

v_grid_lines = np.linspace(x_min, x_max, n_cols + 1)
h_grid_lines = np.linspace(y_min, y_max, m_rows + 1)

ax.vlines(v_grid_lines, y_min, y_max, colors="gray", linestyles="dashed", linewidth=0.5)
ax.hlines(h_grid_lines, x_min, x_max, colors="gray", linestyles="dashed", linewidth=0.5)
ax.set_box_aspect(1)
```
