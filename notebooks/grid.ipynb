```python
%matplotlib inline
import matplotlib.pyplot as plt

from nsfdpy.grid import StaggeredGrid
from nsfdpy.plot import StaggeredGridPlot


geom_data = {
    "imax": 10,
    "jmax": 15,
    "xlength": 100,
    "ylength": 150
}

grid = StaggeredGrid(geom_data)
grid_plot = StaggeredGridPlot(grid)

fig, ax = plt.subplots(1, 1, dpi=200)
plot_boundary = True
s = 5
grid_plot.cells(ax, plot_boundary)
grid_plot.p.scatter(ax, plot_boundary, c="b", marker="$p$", s=s)
grid_plot.u.scatter(ax, plot_boundary, c="g", marker="$u$", s=s)
grid_plot.v.scatter(ax, plot_boundary, c="r", marker="$v$", s=s)
ax.set_aspect('equal', 'box')
```

```python
fig, ax = plt.subplots(1, 1, dpi=200)
plot_boundary = False
s = 5
grid_plot.cells(ax, plot_boundary)
grid_plot.p.scatter(ax, plot_boundary, c="b", marker="$p$", s=s)
grid_plot.u.scatter(ax, plot_boundary, c="g", marker="$u$", s=s)
grid_plot.v.scatter(ax, plot_boundary, c="r", marker="$v$", s=s)
ax.set_aspect('equal', 'box')
```
