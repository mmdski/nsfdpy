```python
%matplotlib inline
import matplotlib.pyplot as plt

from nsfd.grid import Geometry


grid_props = {
    "imax": 10,
    "jmax": 15,
    "xlength": 100,
    "ylength": 150
}

p_grid_geom = Geometry(grid_props, x_frac=0.5, y_frac=0.5)
u_grid_geom = Geometry(grid_props, x_frac=1, y_frac=0.5)
v_grid_geom = Geometry(grid_props, x_frac=0.5, y_frac=1)

fig, ax = plt.subplots(1, 1, dpi=200)
s = 5
p_grid_geom.plot(ax=ax, grid=True, boundary=True, c="b", marker="$p$", s=s)
u_grid_geom.plot(ax=ax, c="g", marker="$u$", s=s)
v_grid_geom.plot(ax=ax, c="r", marker="$v$", s=s)
ax.set_aspect('equal', 'box')
```

```python
from nsfd.grid import plot_boundary_patch, plot_cells_grid

x_coords = []
y_coords = []

for i in range(1, grid_props["imax"] + 1):
    for j in range(1, grid_props["jmax"] + 1):
        x, y = p_grid_geom[i, j]
        x_coords.append(x)
        y_coords.append(y)

fig, ax = plt.subplots(1, 1, dpi=200)
ax.scatter(x_coords, y_coords, s=0.5)
plot_boundary_patch(grid_props, ax)
plot_cells_grid(grid_props, ax)
ax.set_aspect('equal', 'box')
```
